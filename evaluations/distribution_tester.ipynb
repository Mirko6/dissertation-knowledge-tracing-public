```python
import json
import seaborn as sns
from scipy.stats import probplot, norm
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('cross_validation_complete/dkt.json', 'r') as f:
    dkt_results = json.load(f)

with open('cross_validation_complete/saint.json', 'r') as f:
    saint_results = json.load(f)

with open('cross_validation_complete/dkt_plus.json', 'r') as f:
    dkt_plus_results = json.load(f)

with open('cross_validation_complete/akt.json', 'r') as f:
    akt_results = json.load(f)
```

# DKT

```python
sns.kdeplot(dkt_results['auc'])
```

```python
probplot(dkt_results['auc'], dist='norm', plot=plt)
```

# SAINT

```python
sns.kdeplot(saint_results['auc'])
```

```python
probplot(saint_results['auc'], dist='norm', plot=plt)
```

```python
saint_aucs = sorted(saint_results['auc'])
probplot(saint_aucs[2:], dist='norm', plot=plt)
```

# random values

```python
mu, std = norm.fit(dkt_plus_results['auc'])
```

```python
random_aucs = np.random.normal(loc=mu, scale=std, size=10)
random_aucs
probplot(random_aucs, dist='norm', plot=plt)
```
